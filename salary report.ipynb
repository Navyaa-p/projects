```python
x=df_new.drop('salary',axis=1)
y=df_new['salary']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import power_transform
x_new=power_transform(x)
x=pd.DataFrame(x_new,columns=x.columns)
```

```python
#standard scaler to fit the values
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
for i in range (0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=.20)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    print(f'for random state {i},the training accuracy is',{r2_score(y_train,pred_train)})
    print(f'for random state {i} ,the testing accuracy is',{r2_score(y_test, pred_test)})
    print('\n')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=62,test_size=.20)
lr.fit(x_train,y_train)
```

```python
pred_test=lr.predict(x_test)
```

```python
print(r2_score(y_test,pred_test))
```

```python
print('error:')
print('Mean absolute error:',mean_absolute_error(y_test,pred_test))
print('Mean squared error :',mean_squared_error(y_test,pred_test))
print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,pred_test)))
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn= KNeighborsRegressor()
```

```python
for i in range (0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=.20)
    knn.fit(x_train,y_train)
    pred_train=knn.predict(x_train)
    pred_test=knn.predict(x_test)
    print(f'for random state {i},the training accuracy is',{r2_score(y_train,pred_train)})
    print(f'for random state {i} ,the testing accuracy is',{r2_score(y_test, pred_test)})
    print('\n')
```

```python
#ENsemble technique

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

parameters={'criterion':['mse','mae' ],'max_features':['auto','sqrt','log2']}
rf=RandomForestRegressor()
clf=GridSearchCV(rf,parameters)
clf.fit(x_train,y_train)

print(clf.best_params_)
```

```python
rf=RandomForestRegressor(criterion='mae',max_features='sqrt')
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
pred_dec=rf.predict(x_test)

rfs=r2_score(y_test,pred_dec)
print('r2_score:',rfs*100)

rfscore=cross_val_score(rf,x,y,cv=5)
rfc=rfscore.mean()
print('cross val score',rfc*100)
```
